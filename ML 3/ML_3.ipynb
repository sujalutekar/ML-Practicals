```python
import pandas as pd

df = pd.read_csv('/content/Social_Network_Ads.csv')

df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Define features and target
X = df[['Gender', 'Age', 'EstimatedSalary']]
y = df['Purchased']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize KNN classifier with k=5 (you can adjust k as needed)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Make predictions
y_pred = knn.predict(X_test)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Error Rate
error_rate = 1 - accuracy
print("Error Rate:", error_rate)

# Precision
precision = precision_score(y_test, y_pred, zero_division=1)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, y_pred, zero_division=1)
print("Recall:", recall)





# Purchased is chosen as the target variable because it represents the main outcome we want to predict.

# Setting n_neighbors = 5 means that the model will look at the 5 nearest points to the new data point to determine its class.

# test_size = 0.25, here, 0.25 means that 25% of the data will be used for testing, while the remaining 75% will be used for training.

# The random_state parameter controls the randomness of the data splitting. By setting random_state to a fixed integer (like 0), you ensure that the
# data split is reproducible every time you run the code. This means that each time you execute this line of code with the same random_state value, you'll get the
# same train-test split, which is useful for consistency in experiments and comparisons.

```
